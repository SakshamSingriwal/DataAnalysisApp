```python
data=pd.read_excel("sales.xlsx")
```

```python
data
```

```python
result = data.groupby('product').agg(
    newcol=('price','sum')
)

result = result.reset_index()
result
```

```python
result = data.groupby(['state','product']).agg(
    newcol=('price','sum')
)
result = result.reset_index()
result
```
